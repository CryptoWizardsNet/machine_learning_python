# Imports

```python
import pandas as pd
import numpy as np
from pandas_datareader.data import DataReader
import matplotlib.pyplot as plt
```

# Extract and View Data

```python
start_date = "2017-01-01"
end_date = "2022-06-01"
symbol = "ETH-USD"
df = DataReader(name=symbol, data_source='yahoo', start=start_date, end=end_date)
df
```

# Working with Returns

```python
# Add a column
df["Returns"] = df["Close"].pct_change()
df.head(3)
```

```python
# Drop Na
df.dropna(inplace=True)
df.head(3)
```

```python
# Calculate Log Returns
df["lrets"] = np.log(df["Close"] / df["Close"].shift(1))
df.head(3)
```

```python
# CumSum Log Returns
df["lrets_cumsum"] = df["lrets"].cumsum()
df.head()
```

```python
# Normalise Log Returns
df["rets_normal"] = np.exp(df["lrets_cumsum"]) - 1
df.head()
```

```python
# Fill na
df.fillna(0, inplace=True)
df.head(3)
```

# Structure Changes

```python
# Create a new copy
df_new = df.copy()
df_new.head(3)
```

```python
# Drop Columns
df_new.drop(columns=["High", "Low", "Adj Close"], inplace=True)
df_new.head()
```

```python
# Work with certain rows and columns
df_new.iloc[1:3, 2:-1]
```

# Conditionals

```python
# Add conditional statements
df_new.loc[df["Close"].shift(-1) > df["Close"], "TARGET"] = 1
df_new.loc[df["Close"].shift(-1) <= df["Close"], "TARGET"] = -1
df_new.head()
```

# Iterations

```python
# Iterate over DataFrame
i = 0
for index, row in df_new.iterrows():
    print(index, row["Close"], row["TARGET"])
    if i >= 4:
        break
    i += 1
```

# Extract Arrays and Values from DF

```python
# Get Close prices
df_ex = df_new.copy()
close_prices = df_ex["Close"].values
list(close_prices[:4])
```

```python
# Change Close Prices
mod_close_prices = close_prices / 2
mod_close_prices[:4]
```

```python
# Add back to DataFrame
df_ex["Mod_Close"] = mod_close_prices
df_ex.head()
```

```python
# Get just one item
price = df_ex["Close"].iloc[1:2].item()
price
```

# Multiple Adjustments

```python
# Change Close, Volume and Open
df_ex[["Open", "Close", "Volume"]] = df_ex[["Open", "Close", "Volume"]] / df_ex[["Open", "Close", "Volume"]].max()
df_ex.tail()
```
